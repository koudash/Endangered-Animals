```python
# Dependency
import pandas as pd
```

```python
# Read animal species data
df = pd.read_csv("./data/rawdata/animal_species_list.csv", encoding="UTF-8")

# Preview "df"
df.head()
```

```python
# Delete "other common name" column
del df['other common name']

# Preview "df"
df.head()
```

```python
# Check length of "df"
len(df)
```

```python
# Retrieve animal web ID from the full url

# list to store animal web ID
url_ids = []

# Loop through all full length url for animals on www.iucnredlist.org
for i in range(len(df)):
    # Retrieve only animal web ID part of full length url and append it to "url_ids"
    url_id = str(df.url[i].split("/")[4]) + "/" + str(df.url[i].split("/")[5])
    url_ids.append(url_id)
    
# View "url_ids"
url_ids
```

```python
# Add "IUCN_WebID" column in df and save url_ids data
df["IUCN_WebID"] = url_ids

# View "df"
df
```

```python
# Delete "url" column from "df"
del df['url']

# Preview "df"
df.head()
```

```python
# Save df as csv file
df.to_csv("./data/rawdata/animal_species_for_scraping.csv", index=False, header=True)
```
